This file will contain functions to encapsulate logic for making requests to the TMDB API, such as searching for movies and fetching movie details, using the `requests` library and the API key from `config.py`.



````python
import os

movie_api_file_path = os.path.join(project_name, "movie_api.py")

movie_api_content = '''
import requests
from flask import current_app

class MovieAPI:
    BASE_URL = "https://api.themoviedb.org/3"

    def __init__(self):
        self.api_key = current_app.config.get('TMDB_API_KEY')
        if not self.api_key or self.api_key == 'your-tmdb-api-key':
            print("WARNING: TMDB_API_KEY is not set or is a placeholder. API calls will likely fail.")

    def _make_request(self, endpoint, params=None):
        if not self.api_key:
            return None

        url = f"{self.BASE_URL}/{endpoint}"
        headers = {"Authorization": f"Bearer {self.api_key}"}
        # Note: TMDB's new API requires Bearer token, not api_key as a query param for v3/v4.
        # If using v3, some endpoints might still accept 'api_key' in params for backward compatibility.
        # For simplicity and common use, let's stick to Bearer for v4 and most v3 endpoints.
        # If specific v3 endpoints require query 'api_key', this method would need adjustment.

        # For TMDB v3 endpoints like /search/movie or /movie/{movie_id}, the API key is typically
        # passed as a query parameter 'api_key'. Let's adjust for this common use case.
        request_params = {"api_key": self.api_key}
        if params:
            request_params.update(params)

        try:
            response = requests.get(url, params=request_params)
            response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f"API request failed: {e}")
            return None

    def search_movies(self, query, page=1):
        endpoint = "search/movie"
        params = {
            "query": query,
            "page": page
        }
        return self._make_request(endpoint, params)

    def get_movie_details(self, movie_id):
        endpoint = f"movie/{movie_id}"
        return self._make_request(endpoint)

# Initialize the API client globally, but ensure current_app context for config
# We will instantiate this in app.py or blueprints to ensure config is loaded.
'''

with open(movie_api_file_path, "w") as f:
    f.write(movie_api_content)

print(f"Created '{movie_api_file_path}' with TMDB API integration logic.")
````

## Database Schema and Models

### Subtask:
Define the data models for 'Users', 'Movie Ratings', and 'User Watchlists', and implement their persistence logic with Firestore.


**Reasoning**:
I will create a new file named `models.py` inside the `my_flask_app` directory. This file will contain the Python classes for `User`, `MovieRating`, and `UserWatchlist`, along with their respective CRUD methods that interact with Firestore, as per the instructions.



````python
import os
from datetime import datetime

models_file_path = os.path.join(project_name, "models.py")

models_content = '''
from .database import get_db
import datetime

class User:
    COLLECTION_NAME = 'users'

    def __init__(self, id=None, username, email, password_hash, created_at=None):
        self.id = id # Firestore document ID
        self.username = username
        self.email = email
        self.password_hash = password_hash
        self.created_at = created_at if created_at else datetime.datetime.now()

    def to_dict(self):
        return {
            'username': self.username,
            'email': self.email,
            'password_hash': self.password_hash,
            'created_at': self.created_at
        }

    @staticmethod
    def from_dict(user_id, source):
        return User(
            id=user_id,
            username=source['username'],
            email=source['email'],
            password_hash=source['password_hash'],
            created_at=source['created_at']
        )

    @classmethod
    def create_user(cls, username, email, password_hash):
        db = get_db()
        if not db: return None
        user_ref = db.collection(cls.COLLECTION_NAME).document()
        user = cls(id=user_ref.id, username=username, email=email, password_hash=password_hash)
        user_ref.set(user.to_dict())
        return user

    @classmethod
    def get_user_by_email(cls, email):
        db = get_db()
        if not db: return None
        users_ref = db.collection(cls.COLLECTION_NAME)
        query = users_ref.where('email', '==', email).limit(1).get()
        if query:
            user_doc = query[0]
            return cls.from_dict(user_doc.id, user_doc.to_dict())
        return None

    @classmethod
    def get_user_by_id(cls, user_id):
        db = get_db()
        if not db: return None
        user_doc = db.collection(cls.COLLECTION_NAME).document(user_id).get()
        if user_doc.exists:
            return cls.from_dict(user_doc.id, user_doc.to_dict())
        return None

    def update_user(self, updates):
        db = get_db()
        if not db: return False
        user_ref = db.collection(self.COLLECTION_NAME).document(self.id)
        user_ref.update(updates)
        # Update local object attributes
        for key, value in updates.items():
            setattr(self, key, value)
        return True

    def delete_user(self):
        db = get_db()
        if not db: return False
        db.collection(self.COLLECTION_NAME).document(self.id).delete()
        return True

class MovieRating:
    COLLECTION_NAME = 'movie_ratings'

    def __init__(self, id=None, user_id, movie_id, rating, timestamp=None):
        self.id = id # Firestore document ID
        self.user_id = user_id
        self.movie_id = movie_id
        self.rating = rating
        self.timestamp = timestamp if timestamp else datetime.datetime.now()

    def to_dict(self):
        return {
            'user_id': self.user_id,
            'movie_id': self.movie_id,
            'rating': self.rating,
            'timestamp': self.timestamp
        }

    @staticmethod
    def from_dict(rating_id, source):
        return MovieRating(
            id=rating_id,
            user_id=source['user_id'],
            movie_id=source['movie_id'],
            rating=source['rating'],
            timestamp=source['timestamp']
        )

    @classmethod
    def add_rating(cls, user_id, movie_id, rating):
        db = get_db()
        if not db: return None
        rating_ref = db.collection(cls.COLLECTION_NAME).document()
        movie_rating = cls(id=rating_ref.id, user_id=user_id, movie_id=movie_id, rating=rating)
        rating_ref.set(movie_rating.to_dict())
        return movie_rating

    @classmethod
    def get_ratings_by_user(cls, user_id):
        db = get_db()
        if not db: return []
        ratings_ref = db.collection(cls.COLLECTION_NAME)
        query = ratings_ref.where('user_id', '==', user_id).get()
        return [cls.from_dict(doc.id, doc.to_dict()) for doc in query]

    @classmethod
    def get_rating(cls, rating_id):
        db = get_db()
        if not db: return None
        rating_doc = db.collection(cls.COLLECTION_NAME).document(rating_id).get()
        if rating_doc.exists:
            return cls.from_dict(rating_doc.id, rating_doc.to_dict())
        return None

    def update_rating(self, new_rating):
        db = get_db()
        if not db: return False
        rating_ref = db.collection(self.COLLECTION_NAME).document(self.id)
        updates = {'rating': new_rating, 'timestamp': datetime.datetime.now()}
        rating_ref.update(updates)
        self.rating = new_rating
        self.timestamp = updates['timestamp']
        return True

    def delete_rating(self):
        db = get_db()
        if not db: return False
        db.collection(self.COLLECTION_NAME).document(self.id).delete()
        return True

class UserWatchlist:
    COLLECTION_NAME = 'user_watchlists'

    def __init__(self, id=None, user_id, movie_id, added_at=None):
        self.id = id # Firestore document ID
        self.user_id = user_id
        self.movie_id = movie_id
        self.added_at = added_at if added_at else datetime.datetime.now()

    def to_dict(self):
        return {
            'user_id': self.user_id,
            'movie_id': self.movie_id,
            'added_at': self.added_at
        }

    @staticmethod
    def from_dict(watchlist_id, source):
        return UserWatchlist(
            id=watchlist_id,
            user_id=source['user_id'],
            movie_id=source['movie_id'],
            added_at=source['added_at']
        )

    @classmethod
    def add_to_watchlist(cls, user_id, movie_id):
        db = get_db()
        if not db: return None
        # Check if already in watchlist to prevent duplicates
        existing_entry = db.collection(cls.COLLECTION_NAME)
        query = existing_entry.where('user_id', '==', user_id).where('movie_id', '==', movie_id).limit(1).get()
        if query:
            return cls.from_dict(query[0].id, query[0].to_dict())

        watchlist_ref = db.collection(cls.COLLECTION_NAME).document()
        watchlist_item = cls(id=watchlist_ref.id, user_id=user_id, movie_id=movie_id)
        watchlist_ref.set(watchlist_item.to_dict())
        return watchlist_item

    @classmethod
    def get_watchlist_by_user(cls, user_id):
        db = get_db()
        if not db: return []
        watchlist_ref = db.collection(cls.COLLECTION_NAME)
        query = watchlist_ref.where('user_id', '==', user_id).get()
        return [cls.from_dict(doc.id, doc.to_dict()) for doc in query]

    def remove_from_watchlist(self):
        db = get_db()
        if not db: return False
        db.collection(self.COLLECTION_NAME).document(self.id).delete()
        return True
'''

with open(models_file_path, "w") as f:
    f.write(models_content)

print(f"Created '{models_file_path}' with User, MovieRating, and UserWatchlist models.")
````

````python
import os
import jwt
import datetime
from functools import wraps

from flask import Flask, jsonify, request, g
from werkzeug.security import generate_password_hash, check_password_hash

from .database import get_db
from .config import Config
from .movie_api import MovieAPI
from .models import User

app = Flask(__name__)
app.config.from_object(Config)

db = get_db()
movie_api = MovieAPI() # Instantiate MovieAPI here

# --- JWT Authentication Decorator ---
def token_required(f):
    @wraps(f)
    def decorated(*args, **kwargs):
        token = None

        if 'Authorization' in request.headers:
            token = request.headers['Authorization'].split(" ")[1]

        if not token:
            return jsonify({'message': 'Token is missing!'}), 401

        try:
            data = jwt.decode(token, app.config['JWT_SECRET_KEY'], algorithms=["HS256"])
            current_user = User.get_user_by_id(data['user_id'])
            if current_user is None:
                return jsonify({'message': 'Token is invalid or user not found!'}), 401
            g.current_user = current_user # Store current user in global context
        except jwt.ExpiredSignatureError:
            return jsonify({'message': 'Token has expired!'}), 401
        except jwt.InvalidTokenError:
            return jsonify({'message': 'Token is invalid!'}), 401

        return f(*args, **kwargs)
    return decorated

# --- Existing Routes ---
@app.route('/')
def hello_world():
    return 'Hello, World! This is our movie recommendation app.'

@app.route('/test_db')
def test_db():
    if db:
        try:
            # Try to get a collection reference to check connectivity
            db.collection('test_collection').document('test_doc').get()
            return 'Successfully connected to Firestore!'
        except Exception as e:
            return f'Failed to connect to Firestore: {e}'
    else:
        return 'Firestore database not initialized.'

@app.route('/search_movies')
def search_movies_route():
    query = request.args.get('query')
    if not query:
        return jsonify({'error': 'Query parameter "query" is required'}), 400

    results = movie_api.search_movies(query)
    if results:
        return jsonify(results)
    else:
        return jsonify({'error': 'Could not fetch movies from TMDB API'}), 500

# --- New User Authentication Routes ---
@app.route('/register', methods=['POST'])
def register_user():
    data = request.get_json()
    if not data or not all(key in data for key in ['username', 'email', 'password']):
        return jsonify({'message': 'Missing username, email, or password'}), 400

    username = data['username']
    email = data['email']
    password = data['password']

    if User.get_user_by_email(email):
        return jsonify({'message': 'User with this email already exists'}), 409

    hashed_password = generate_password_hash(password, method='scrypt') # Use 'scrypt' as a modern hashing method
    new_user = User.create_user(username=username, email=email, password_hash=hashed_password)

    if new_user:
        return jsonify({'message': 'User registered successfully!'}), 201
    else:
        return jsonify({'message': 'Failed to register user.'}), 500

@app.route('/login', methods=['POST'])
def login_user():
    data = request.get_json()
    if not data or not all(key in data for key in ['email', 'password']):
        return jsonify({'message': 'Missing email or password'}), 400

    email = data['email']
    password = data['password']

    user = User.get_user_by_email(email)
    if not user or not check_password_hash(user.password_hash, password):
        return jsonify({'message': 'Invalid credentials!'}), 401

    # Generate JWT token
    token = jwt.encode({
        'user_id': user.id,
        'exp': datetime.datetime.utcnow() + datetime.timedelta(minutes=30) # Token expires in 30 minutes
    }, app.config['JWT_SECRET_KEY'], algorithm="HS256")

    return jsonify({'message': 'Login successful!', 'token': token}), 200

# Example of a protected route
@app.route('/protected', methods=['GET'])
@token_required
def protected_route():
    # Access the current user from g.current_user
    return jsonify({'message': f'Hello, {g.current_user.username}! You accessed a protected route.'})


if __name__ == '__main__':
    app.run(debug=True)
````

## Recommendation Engine Implementation

### Subtask:
Implement a basic movie recommendation algorithm, such as a content-based approach based on user ratings and movie genres.


**Reasoning**:
I will create a new file named `recommendation_engine.py` inside the `my_flask_app` directory and define the `RecommendationEngine` class with the `get_recommendations_for_user` method skeleton, as instructed.



````python
import os

recommendation_engine_file_path = os.path.join(project_name, "recommendation_engine.py")

recommendation_engine_content = '''
from .database import get_db
from .models import MovieRating, UserWatchlist # UserWatchlist is added for future exclusion
from .movie_api import MovieAPI

class RecommendationEngine:
    def __init__(self):
        self.db = get_db()
        self.movie_api = MovieAPI()

    def get_recommendations_for_user(self, user_id, num_recommendations=10):
        if not self.db:
            print("Firestore DB not initialized.")
            return []

        # 1. Retrieve the movies that the specified user has rated
        user_ratings = MovieRating.get_ratings_by_user(user_id)

        # 2. Filter for movies with a high rating (e.g., >= 4 stars)
        highly_rated_movie_ids = [r.movie_id for r in user_ratings if r.rating >= 4]
        if not highly_rated_movie_ids:
            print(f"User {user_id} has no highly-rated movies (>= 4 stars).")
            return []

        # Collect genres from highly-rated movies
        user_preferred_genres = set()
        processed_movie_ids = set()

        print(f"Fetching details for {len(highly_rated_movie_ids)} highly rated movies...")
        for movie_id in highly_rated_movie_ids:
            if movie_id in processed_movie_ids:
                continue
            processed_movie_ids.add(movie_id)

            movie_details = self.movie_api.get_movie_details(movie_id)
            if movie_details and 'genres' in movie_details:
                for genre in movie_details['genres']:
                    user_preferred_genres.add(genre['id']) # TMDB genre IDs

        if not user_preferred_genres:
            print("Could not determine preferred genres from highly rated movies.")
            return []

        print(f"User preferred genres (IDs): {list(user_preferred_genres)}")

        # 5. Use the TMDB API's discovery or search features to find movies that match these genres
        # TMDB Discover API is suitable for this. genre IDs are comma separated.
        genre_ids_str = ','.join(map(str, user_preferred_genres))
        if not genre_ids_str:
            return []

        # Exclude movies user has already rated
        rated_movie_ids = {r.movie_id for r in user_ratings}
        # Also exclude movies in user's watchlist (if implemented and desired)
        user_watchlist = UserWatchlist.get_watchlist_by_user(user_id)
        watchlist_movie_ids = {item.movie_id for item in user_watchlist}
        excluded_movie_ids = rated_movie_ids.union(watchlist_movie_ids)

        # TMDB Discover API: sort by popularity, filter by genres
        # Note: TMDB Discover does not have a direct 'exclude_movie_ids' parameter.
        # We will need to filter locally after getting results.
        # The `with_genres` parameter takes a comma-separated list of genre IDs.
        discovery_params = {
            "with_genres": genre_ids_str,
            "sort_by": "popularity.desc",
            "vote_count.gte": 50, # Only movies with a reasonable number of votes
            "page": 1
        }

        recommended_movies = []
        pages_to_check = 3 # Check a few pages to get enough recommendations

        print(f"Searching for movies with genres: {genre_ids_str}")
        for page in range(1, pages_to_check + 1):
            discovery_params["page"] = page
            discover_results = self.movie_api._make_request("discover/movie", discovery_params)
            if discover_results and 'results' in discover_results:
                for movie in discover_results['results']:
                    # Filter out movies already rated or in watchlist
                    if str(movie['id']) not in excluded_movie_ids:
                        recommended_movies.append(movie)
                        if len(recommended_movies) >= num_recommendations:
                            break
            if len(recommended_movies) >= num_recommendations:
                break
            if not discover_results or discover_results['page'] >= discover_results['total_pages']:
                break

        print(f"Found {len(recommended_movies)} recommendations for user {user_id}.")
        return recommended_movies[:num_recommendations]
'''

with open(recommendation_engine_file_path, "w") as f:
    f.write(recommendation_engine_content)

print(f"Created '{recommendation_engine_file_path}' with RecommendationEngine class and method skeleton.")
````

````python
import os

recommendation_engine_file_path = os.path.join(project_name, "recommendation_engine.py")

# Updated content for recommendation_engine.py with added comments
recommendation_engine_content_commented = '''
from database import get_db # Direct import for compatibility
from models import MovieRating, UserWatchlist # UserWatchlist is added for future exclusion in recommendations
from movie_api import MovieAPI

# RecommendationEngine class implements a content-based recommendation algorithm.
# It provides personalized movie recommendations based on a user's highly-rated movies.
class RecommendationEngine:
    # Initializes the RecommendationEngine with database and movie API clients.
    def __init__(self):
        self.db = get_db()
        self.movie_api = MovieAPI()

    # Generates a list of movie recommendations for a given user.
    # user_id: The ID of the user for whom to generate recommendations.
    # num_recommendations: The desired number of recommendations to return (default: 10).
    # Returns a list of dictionaries, each representing a recommended movie.
    def get_recommendations_for_user(self, user_id, num_recommendations=10):
        if not self.db:
            print("Firestore DB not initialized.")
            return []

        # 1. Retrieve all movies that the specified user has rated.
        user_ratings = MovieRating.get_ratings_by_user(user_id)

        # 2. Filter for movies with a high rating (e.g., >= 4 stars) to identify user preferences.
        highly_rated_movie_ids = [r.movie_id for r in user_ratings if r.rating >= 4]
        if not highly_rated_movie_ids:
            print(f"User {user_id} has no highly-rated movies (>= 4 stars).")
            return []

        # 3. Collect genres from these highly-rated movies to determine user's preferred genres.
        user_preferred_genres = set()
        processed_movie_ids = set() # To avoid redundant API calls for movie details

        print(f"Fetching details for {len(highly_rated_movie_ids)} highly rated movies...")
        for movie_id in highly_rated_movie_ids:
            if movie_id in processed_movie_ids:
                continue
            processed_movie_ids.add(movie_id)

            # Fetch movie details from TMDB API to get genre information
            movie_details = self.movie_api.get_movie_details(movie_id)
            if movie_details and 'genres' in movie_details:
                for genre in movie_details['genres']:
                    user_preferred_genres.add(genre['id']) # TMDB genre IDs are integers

        if not user_preferred_genres:
            print("Could not determine preferred genres from highly rated movies.")
            return []

        print(f"User preferred genres (IDs): {list(user_preferred_genres)}")

        # 4. Prepare list of movie IDs to exclude from recommendations.
        # This includes movies the user has already rated and those in their watchlist.
        rated_movie_ids = {r.movie_id for r in user_ratings}
        user_watchlist = UserWatchlist.get_watchlist_by_user(user_id)
        watchlist_movie_ids = {item.movie_id for item in user_watchlist}
        excluded_movie_ids = rated_movie_ids.union(watchlist_movie_ids)

        # 5. Use the TMDB Discover API to find movies that match these genres.
        # The 'with_genres' parameter takes a comma-separated list of genre IDs.
        genre_ids_str = ','.join(map(str, user_preferred_genres))
        if not genre_ids_str:
            return []

        discovery_params = {
            "with_genres": genre_ids_str,
            "sort_by": "popularity.desc", # Sort by popularity to get good recommendations
            "vote_count.gte": 50, # Only consider movies with a reasonable number of votes
            "page": 1
        }

        recommended_movies = []
        pages_to_check = 3 # Check multiple pages of discovery results to gather enough recommendations

        print(f"Searching for movies with genres: {genre_ids_str}")
        for page in range(1, pages_to_check + 1):
            discovery_params["page"] = page
            discover_results = self.movie_api._make_request("discover/movie", discovery_params)
            if discover_results and 'results' in discover_results:
                for movie in discover_results['results']:
                    # Filter out movies that the user has already rated or added to their watchlist.
                    # Note: TMDB movie IDs are typically integers, but stored as strings in our DB, so convert for comparison.
                    if str(movie['id']) not in excluded_movie_ids:
                        recommended_movies.append(movie)
                        if len(recommended_movies) >= num_recommendations:
                            break # Stop once enough recommendations are found
            if len(recommended_movies) >= num_recommendations:
                break
            # Stop if there are no more pages or results
            if not discover_results or discover_results['page'] >= discover_results['total_pages']:
                break

        print(f"Found {len(recommended_movies)} recommendations for user {user_id}.")
        # Return the requested number of recommendations
        return recommended_movies[:num_recommendations]
'''

with open(recommendation_engine_file_path, "w") as f:
    f.write(recommendation_engine_content_commented)

print(f"Updated '{recommendation_engine_file_path}' with detailed comments.")
````
